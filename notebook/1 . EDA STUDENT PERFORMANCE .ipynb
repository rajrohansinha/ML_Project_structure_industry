### 1) Problem Statement
- The goal of this project is to analyze how various factors such as gender, ethnicity, parental education level, lunch type, and test preparation course influence students' test scores.

### 2) Data Collection
- Dataset Source: [Kaggle - Students Performance in Exams](https://www.kaggle.com/datasets/spscientist/students-performance-in-exams?datasetId=74977)
- The dataset contains 1000 rows and 8 columns.

### 3.8 Adding columns for "Total Score" and "Average"

```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['average'] = df['total score']/3
df.head()
```

```python
reading_full = df[df['reading_score'] == 100]['average'].count()
writing_full = df[df['writing_score'] == 100]['average'].count()
math_full = df[df['math_score'] == 100]['average'].count()

print(f'Number of students with full marks in Maths: {math_full}')
print(f'Number of students with full marks in Writing: {writing_full}')
print(f'Number of students with full marks in Reading: {reading_full}')
```

```python
reading_less_20 = df[df['reading_score'] <= 20]['average'].count()
writing_less_20 = df[df['writing_score'] <= 20]['average'].count()
math_less_20 = df[df['math_score'] <= 20]['average'].count()

print(f'Number of students with less than 20 marks in Maths: {math_less_20}')
print(f'Number of students with less than 20 marks in Writing: {writing_less_20}')
print(f'Number of students with less than 20 marks in Reading: {reading_less_20}')
```

#####  Insights
 - From above values we get students have performed the worst in Maths 
 - Best performance is in reading section

#### 4.3 Multivariate analysis using pieplot

```python
plt.figure(figsize=(30, 8))
for i, col in enumerate(categorical_features, 1):
    plt.subplot(1, len(categorical_features), i)
    sizes = df[col].value_counts()
    explode = [0.05] * len(sizes)  # Slightly separate each slice
    plt.pie(
        sizes,
        labels=sizes.index,
        autopct='%1.1f%%',
        startangle=90,
        shadow=True,
        explode=explode,
        textprops={'fontsize': 16}
    )
    plt.title(col.replace('_', ' ').title(), fontsize=20)
    plt.axis('equal')
plt.tight_layout()
plt.show()
```

#####  Insights
- Number of Male and Female students is almost equal
- Number students are greatest in Group C
- Number of students who have standard lunch are greater
- Number of students who have not enrolled in any test preparation course is greater
- Number of students whose parental education is "Some College" is greater followed closely by "Associate's Degree"

#### BIVARIATE ANALYSIS (  Is lunch type intake has any impact on student's performance ? )

```python
#  Compare Average Scores by Parental Education Level, Lunch & Test Prep

#  1. Group by parental education + lunch + test prep
score_group = df.groupby(['parental_level_of_education', 'lunch', 'test_preparation_course'])[['math_score', 'reading_score', 'writing_score', 'average']].mean().reset_index()

#  2. Plot Setup
f, ax = plt.subplots(1, 2, figsize=(22, 10))

#  Plot 1: Average Score by Parental Education & Lunch
sns.barplot(
    x='parental_level_of_education', 
    y='average',
    hue='lunch', 
    data=score_group, 
    palette='coolwarm', 
    saturation=0.9,
    ax=ax[0]
)
ax[0].set_title(' Average Student Score vs Parental Education (Lunch Type)', fontsize=22, fontweight='bold')
ax[0].set_xlabel('Parental Education Level', fontsize=16)
ax[0].set_ylabel('Average Score', fontsize=16)
ax[0].tick_params(axis='x', rotation=30)

for container in ax[0].containers:
    ax[0].bar_label(container, fontsize=14, color='black', weight='bold', fmt='%.1f')

#  Plot 2: Average Score by Parental Education & Test Prep
sns.barplot(
    x='parental_level_of_education', 
    y='average',
    hue='test_preparation_course', 
    data=score_group, 
    palette='viridis', 
    saturation=0.9,
    ax=ax[1]
)
ax[1].set_title(' Average Student Score vs Parental Education (Test Prep)', fontsize=22, fontweight='bold')
ax[1].set_xlabel('Parental Education Level', fontsize=16)
ax[1].set_ylabel('Average Score', fontsize=16)
ax[1].tick_params(axis='x', rotation=30)

for container in ax[1].containers:
    ax[1].bar_label(container, fontsize=14, color='black', weight='bold', fmt='%.1f')

plt.tight_layout()
plt.show()
```

#### Insights 
- Students who get Standard Lunch tend to perform better than students who got free/reduced lunch
